```python
#подключаем библиотеку
import pandas as pd
#путь и читаем
iowa_file_path = 'train.csv'
home_data = pd.read_csv(iowa_file_path)
```

сначала выберем параметр, которым будем управлять. назовем его y

```python
home_data
```

```python
y = home_data.SalePrice
```

теперь создадим DataFrame, в котором будут аргументы функции

```python

feature_names = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF',
                 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']

X = home_data[feature_names]
```

посмотрим на нее

выбираем модель и помещаем ее

```python
from sklearn.tree import DecisionTreeRegressor

iowa_model = DecisionTreeRegressor(random_state=1)

iowa_model.fit(X, y)
```

предсказываем результат согласно модели

```python
predictions = iowa_model.predict(X)
print(predictions)
y.head()
```

данная модель близка, но не является точной

Вам захочется оценить почти каждую модель, которую вы когда-либо создавали. В большинстве (хотя и не во всех) приложениях значимым показателем качества модели является точность прогнозирования. Другими словами, будут ли предсказания модели близки к тому, что происходит на самом деле.

Многие люди допускают огромную ошибку при измерении точности прогнозирования. Они делают прогнозы на основе своих обучающих данных и сравнивают эти прогнозы с целевыми значениями в обучающих данных. Вы сразу увидите проблему этого подхода и способы ее решения, но давайте сначала подумаем, как бы мы это сделали.

Существует множество показателей для обобщения качества модели, но мы начнем с показателя, называемого средней абсолютной ошибкой (также называемого MAE).

разделим наши данные

```python

from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
```

создадим новую модель по поделенным данным

```python
iowa_model = DecisionTreeRegressor(random_state=1)

iowa_model.fit(train_X, train_y)
```

делаем предположение

```python
val_predictions = iowa_model.predict(val_X)

print(val_predictions)
```

вычислим ошибку

```python
from sklearn.metrics import mean_absolute_error
val_mae = mean_absolute_error(val_y, val_predictions)

print(val_mae)
```

ужасно, ниже мы ее улучшим

Когда мы разделяем дома на множество листьев, у нас также становится меньше домов на каждом листе. Листья с очень небольшим количеством домов будут давать прогнозы, которые очень близки к фактическим значениям этих домов, но они могут давать очень ненадежные прогнозы для новых данных (поскольку каждый прогноз основан только на нескольких домах).

Это явление, называемое переоснащением, когда модель почти идеально соответствует обучающим данным, но плохо справляется с проверкой и другими новыми данными. С другой стороны, если мы сделаем наше дерево очень мелким, оно не будет делить дома на отдельные группы.

В крайнем случае, если дерево делит дома всего на 2 или 4, в каждой группе все равно будет большое разнообразие домов. Полученные прогнозы могут быть далекими для большинства домов, даже в обучающих данных (и они будут плохими при проверке по той же причине). Когда модель не может уловить важные различия и закономерности в данных, поэтому она плохо работает даже на обучающих данных, это называется недостаточной подгонкой.

Поскольку нас заботит точность новых данных, которые мы оцениваем на основе наших данных проверки, мы хотим найти золотую середину между недостаточным и переоснащением. Визуально нам нужна нижняя точка (красной) кривой проверки на рисунке ниже.

напишем функциЮ, которая считает ошибку по глубине дерева

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

найдем лучшее число

```python
candidate_max_leaf_nodes = [5, 25, 50, 100, 250, 500]
a = []

for max_leaf_nodes in candidate_max_leaf_nodes:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    a.append([max_leaf_nodes, my_mae])

# минимальное по второму элементу в списке
best_tree_size = min(a, key=lambda x: x[1])[0]
print(min(a, key=lambda x: x[1]))
```

поместим модель

```python

final_model = DecisionTreeRegressor(max_leaf_nodes=best_tree_size, random_state=0)


final_model.fit(X, y)
```

Это хорошо, МАЭ? Не существует общего правила, определяющего, какие значения являются хорошими и применимого ко всем приложениям. Но на следующем шаге вы увидите, как использовать (и улучшить) это число.

Даже самые сложные методы моделирования сегодня сталкиваются с противоречием между недостаточным и переоснащением. Но во многих моделях есть умные идеи, которые могут привести к повышению производительности. В качестве примера мы рассмотрим случайный лес.

Случайный лес использует множество деревьев и делает прогноз путем усреднения прогнозов каждого компонента дерева. Обычно он имеет гораздо лучшую точность прогнозирования, чем одно дерево решений, и хорошо работает с параметрами по умолчанию. Если вы продолжите моделировать, вы сможете изучить больше моделей с еще большей производительностью, но многие из них чувствительны к получению правильных параметров.

Мы строим модель случайного леса аналогично тому, как мы строили дерево решений в scikit-learn — на этот раз используя класс RandomForestRegressor вместо DecisionTreeRegressor.

```python
from sklearn.ensemble import RandomForestRegressor

# Define the model. Set random_state to 1
rf_model = RandomForestRegressor(random_state=1)

# fit your model
rf_model.fit(train_X, train_y)

# Calculate the mean absolute error of your Random Forest model on the validation data
rf_val_mae = mean_absolute_error(val_y, rf_model.predict(val_X))

print("Validation MAE for Random Forest Model: {}".format(rf_val_mae))
```
